```python
import numpy
import pandas
import matplotlib.pyplot as plt
```

```python
date_k2c5 = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 17, max_rows = 3378, usecols = 2)
flux_k2c5 = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 17, max_rows = 3378, usecols = 3)
date_k2c18 = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 3395, max_rows = 2265, usecols = 2)
flux_k2c18 = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 3395, max_rows = 2265, usecols = 3)
date_hat = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 5660, max_rows = 12903, usecols = 2)
flux_hat = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 5660, max_rows = 12903, usecols = 3)
date_kelt = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 18562, max_rows = 4690, usecols = 2)
flux_kelt = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 18562, max_rows = 4690, usecols = 3)
date_wasp = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 23251, usecols = 2)
flux_wasp = numpy.genfromtxt("becker2018b_hip41378_rawdata.csv", delimiter = ",", skip_header = 23251, usecols = 3)
```
